# Linear Regression - House Prices

### Preprocessing

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='darkgrid')
```

```python
df = pd.read_csv('https://github.com/codebasics/py/raw/master/ML/1_linear_reg/homeprices.csv')
df.head()
```

```python
df.describe()
```

```python
df.plot.scatter('area', 'price', color='red')
plt.show()
```

```python
for x in df.columns:
    df[x].plot.density()
    plt.show()
```

```python
X_train = df.drop('price', axis=1)
y_train = df['price']
```

### Modelling

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_, model.intercept_
```

```python
df.plot.scatter('area', 'price', color='blue')
plt.plot(
    np.array(df['area']),
    model.coef_[0] * np.array(df['area']) + model.intercept_,
    color='red'
)
plt.show()
```

### Evaluation

```python
test_df = pd.read_csv('https://github.com/codebasics/py/raw/master/ML/1_linear_reg/areas.csv')
test_df.head()
```

```python
test_df['area'].plot.density()
plt.show()
```

```python
test_df['price'] = model.predict(test_df)
test_df
```
